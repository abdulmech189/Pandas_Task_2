###### Department wise toppers (horizantal bar graph or any visual representations of your choice)

```python
y = df['CodeKata Score'].groupby(df['Department'])
```

```python
toppers = y.nlargest(1)
print(toppers)
```

```python
toppers_index = []
for i in range(len(toppers)):
  toppers_index.append(toppers.index[i][1])
```

```python
df_toppers = df.loc[toppers_index, ['Name', 'Department', 'CodeKata Score']].reset_index(drop = True)
df_toppers
```

```python
for i in range(len(df_toppers)):
  plt.barh(df_toppers.loc[i, 'Name'], df_toppers.loc[i, 'CodeKata Score'], label = df_toppers.loc[i, 'Department'])
  plt.legend()
  plt.xlabel('Codekata score')
  plt.ylabel('Toppers (each department)')
```
